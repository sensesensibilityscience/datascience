```python
#Just run this cell
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

The next few parts will involve filling in code.

```python
#Assigning each station's data to it's own dataframe
basel_df = pd.read_csv("data/temp_data_basel.csv").iloc[:, 1:]
bern_df = pd.read_csv("data/temp_data_bern.csv").iloc[:, 1:]
geneva_df = pd.read_csv("data/temp_data_geneva.csv").iloc[:, 1:]
zurich_df = pd.read_csv("data/temp_data_zurich.csv").iloc[:, 1:]
```

```python
#Aggregating data of all stations, filtering for July values, grouping by year across stations, collecting mean values
combined_df = pd.concat([basel_df, bern_df, geneva_df, zurich_df], axis=0)
june_combined_df = combined_df[(combined_df["Month"] == 6)].iloc[:, [1,3, 4]] #June
july_combined_df = combined_df[(combined_df["Month"] == 7)].iloc[:, [1,3, 4]] #July
aug_combined_df = combined_df[(combined_df["Month"] == 8)].iloc[:, [1,3, 4]] #August
jja_combined_df = combined_df[(combined_df["Month"] == 6) | (combined_df["Month"] == 7) | (combined_df["Month"] == 8)].iloc[:, [1,3, 4]] #JJA
jja_mean_temps = jja_combined_df.groupby("Year").agg("mean").reset_index()
june_mean_temps = june_combined_df.groupby("Year").agg("mean").reset_index()
july_mean_temps = july_combined_df.groupby("Year").agg("mean").reset_index()
aug_mean_temps = aug_combined_df.groupby("Year").agg("mean").reset_index()


#Data from 1864-2003 (as done in study)
study_mean_temps = jja_mean_temps[jja_mean_temps["Year"] < 2004]
june_mean_temps = june_mean_temps[june_mean_temps["Year"] < 2004]
july_mean_temps = july_mean_temps[july_mean_temps["Year"] < 2004]
aug_mean_temps = aug_mean_temps[aug_mean_temps["Year"] < 2004]
```

```python
#Example dataset students will work with
jja_mean_temps.head(5)
```

```python
#Recreating the plot from study, 2003 summer is shown in red (on the right)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(8, 12))

sns.rugplot(june_mean_temps["Temperature"], color='blue', ax=ax1)
sns.rugplot(june_mean_temps[june_mean_temps["Year"] == 2003]["Temperature"], color='red', label='2003', ax=ax1)
sns.histplot(june_mean_temps["Temperature"], color='pink', bins = 20, kde=True, ax=ax1)  # Adjust bins here
ax1.set_title('Historical Range of June Temperatures')
ax1.set_xlabel('Temperature (C)')
ax1.set_ylabel('Frequency')
ax1.legend()

sns.rugplot(july_mean_temps["Temperature"], color='blue', ax=ax2)
sns.rugplot(july_mean_temps[july_mean_temps["Year"] == 2003]["Temperature"], color='red', label='2003', ax=ax2)
sns.histplot(july_mean_temps["Temperature"], color='pink', bins = 20, kde=True, ax=ax2)  # Adjust bins here
ax2.set_title('Historical Range of July Temperatures')
ax2.set_xlabel('Temperature (C)')
ax2.set_ylabel('Frequency')
ax2.legend()

sns.rugplot(aug_mean_temps["Temperature"], color='blue', ax=ax3)
sns.rugplot(aug_mean_temps[aug_mean_temps["Year"] == 2003]["Temperature"], color='red', label='2003', ax=ax3)
sns.histplot(aug_mean_temps["Temperature"], color='pink', bins = 20, kde=True, ax=ax3)  # Adjust bins here
ax3.set_title('Historical Range of August Temperatures')
ax3.set_xlabel('Temperature (C)')
ax3.set_ylabel('Frequency')
ax3.legend()


sns.rugplot(study_mean_temps["Temperature"], color='blue', ax=ax4)
sns.rugplot(study_mean_temps[study_mean_temps["Year"] == 2003]["Temperature"], color='red', label='2003', ax=ax4)
sns.histplot(study_mean_temps["Temperature"], color='pink', bins = 20, kde=True, ax=ax4)  # Adjust bins here
ax4.set_title('Historical Range of Summer Temperatures')
ax4.set_xlabel('Temperature (C)')
ax4.set_ylabel('Frequency')
ax4.legend()

plt.subplots_adjust(hspace=0.5) 
```
